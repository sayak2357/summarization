```python
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import operator
import os
```

```python
directory = 'acl'
pp=[]
x=[]
for i in range(79,99):
    directory = 'acl/'+'acl'+str(i)+'/'
    x.append(i)
    paper=0
    for filename in os.listdir(directory):
        paper+=1
    pp.append(paper)



for i in range(0,19):
    directory = 'acl/'+'acl'+str(i)+'/'
    x.append(i)
    paper=0
    for filename in os.listdir(directory):
        paper+=1
    pp.append(paper)
    
```

```python
print(pp)
print(x)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects=x
y_pos = np.arange(len(objects))
performance = pp
plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('url count')
plt.xlabel('year (2000+) ')
plt.title('number of papers per year of ACL')
 
plt.show()
```
