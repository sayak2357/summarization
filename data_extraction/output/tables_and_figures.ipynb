# ACL

```python
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import operator
import re
import glob
```

```python
dict1 = {}
for i in range(79,100):
    n1 =str("output/")+str("P")+str(i)+str("*")+str("tab_figure_caption*")
    tp=0
    avgtft=0
    for name in glob.glob(n1):
        print(name)
        tp+=1
        t=open(name,'r')
        try:
            soup1=BeautifulSoup(t,'html.parser')
            result1=soup1.find_all('figure')
            result2=soup1.find_all('table')
            tft= len(result1) + len(result2)
            avgtft += tft
            print(len(result1),len(result2),tp,avgtft)
        except:
            continue
    yavg= avgtft /tp
    dict1[i]=yavg
```

```python
print(dict1)
```

```python
for i in range(79,100):
    print(dict1[i], end =" ")
```

```python
dict1 = {}
for i in range(0,19):
    if i==0:
        n1 =str("output/")+str("A0")+str(i)+str("*")+str("tab_figure_caption*")
    elif i<10:
        n1 =str("output/")+str("N0")+str(i)+str("*")+str("tab_figure_caption*")
    else:
        n1 =str("output/")+str("N")+str(i)+str("*")+str("tab_figure_caption*")
    tp=0
    avgtft=0
    for name in glob.glob(n1):
        print(name)
        tp+=1
        t=open(name,'r')
        try:
            soup1=BeautifulSoup(t,'html.parser')
            result1=soup1.find_all('figure')
            result2=soup1.find_all('table')
            tft= len(result1) + len(result2)
            avgtft += tft
            print(len(result1),len(result2),tp,avgtft)
        except:
            continue
    if tp==0:
        tp=1
    yavg= avgtft /tp
    dict1[i]=yavg
```

```python
print(dict1)
```

```python
for i in range(0,19):
    print(dict1[i], end = " ")
```
