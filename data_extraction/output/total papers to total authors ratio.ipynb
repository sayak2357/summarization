# ACL Papers per Author

```python
from bs4 import BeautifulSoup
```

```python
m=[]
n=[]
l=[]
```

```python
for j in range(79,100):
    n1='acl/acl'+str(j)+'.txt'
    f1=open(n1,errors='ignore')
    soup1=BeautifulSoup(f1,'html.parser')
    result1=soup1.find_all('author')
    result2=soup1.find_all('paper')
    unique_name={}
    papers=len(result2)
    authors=len(result1)
    unique_authors={}
    
    for k in range(0,len(result1)):
        fname=result1[k].firstname.text
        sname=result1[k].secondname.text
        uname=fname+sname
        if uname in unique_authors:
            a=unique_authors[uname]
            a=a+1
            unique_authors[uname]=a
        else:
            unique_authors[uname]=1
    #print(authors,len(unique_authors))
    papersbyauthors=papers/authors
    uniauth= len(unique_authors)
    papersbyuniqauth= papers/uniauth
    #print(papersbyauthors)
    m.append(j)
    n.append(papersbyauthors)
    l.append(papersbyuniqauth)
    #print(m,n)
f1.close()
```

```python
for j in range(0,19):
    n1='acl/acl'+str(j)+'.txt'
    f1=open(n1,errors='ignore')
    soup1=BeautifulSoup(f1,'html.parser')
    result1=soup1.find_all('author')
    result2=soup1.find_all('paper')
    unique_name={}
    papers=len(result2)
    authors=len(result1)
    unique_authors={}
    for k in range(0,len(result1)):
        fname=result1[k].firstname.text
        sname=result1[k].secondname.text
        uname=fname+sname
        if uname in unique_authors:
            a=unique_authors[uname]
            a=a+1
            unique_authors[uname]=a
        else:
            unique_authors[uname]=1
    papersbyauthors=papers/authors
    #print(papersbyauthors)
    m.append(j)
    n.append(papersbyauthors)
    #print(m,n)
f1.close()
```

```python
print(n)
```

```python
print(m)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects=m
y_pos = np.arange(len(objects))
performance = n
plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('total papers by total authors')
plt.xlabel('year')
plt.title('ACL total paper to total authors ratio')
 
plt.show()
```
