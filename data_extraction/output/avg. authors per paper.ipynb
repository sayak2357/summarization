```python
from bs4 import BeautifulSoup
```

```python
m=[]
n=[]
```

```python
for j in range(79,100):
    n1='acl/acl'+str(j)+'.txt'
    f1=open(n1,errors='ignore')
    soup1=BeautifulSoup(f1,'html.parser')
    result1=soup1.find_all('author')
    result2=soup1.find_all('paper')
    unique_name={}
    papers=len(result2)
    authors=len(result1)
    
    avgauthors= authors/papers
    #print(papersbyauthors)
    m.append(j)
    n.append(avgauthors)
     #print(m,n)
f1.close()
```

```python
for j in range(0,19):
    n1='acl/acl'+str(j)+'.txt'
    f1=open(n1,errors='ignore')
    soup1=BeautifulSoup(f1,'html.parser')
    result1=soup1.find_all('author')
    result2=soup1.find_all('paper')
    unique_name={}
    papers=len(result2)
    authors=len(result1)

    avgauthors= authors/papers
    #print(papersbyauthors)
    if j<10:
        j='0'+str(j)
    m.append(j)
    n.append(avgauthors)
    #print(m,n)
f1.close()
```

```python
print(m)
```

```python
print(n)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects=m
y_pos = np.arange(len(objects))
performance = n
plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('total authors to total papers ratio')
plt.xlabel('year')
plt.title('ACL avg. no. of authors per paper')
 
plt.show()
```
