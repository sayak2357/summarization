```python
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import operator
import re
```

# Scanning footnote.xml for URLs

```python
fd={}
with open("footnote.txt") as file:
        for line in file:
            urls = re.findall('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+', line)
            if len(urls)>0:
                dt=urls[0]
                if dt in fd:
                    fd[dt]+=1
                else:
                    fd[dt]=1
```

# Scanning url.xml for URLs

```python
ud={}
with open("urls.txt") as file:
        for line in file:
            url = re.findall('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+', line)
            if len(url)>0:
                d=url[0]
                if d in ud:
                    ud[d]+=1
                else:
                    ud[d]=1
```

# URLS present both in Footnote.xml and URL.xml

```python
ct=0
for x in fd:
    for y in ud:
        if x==y:
            print(x)
            ct+=1
```
