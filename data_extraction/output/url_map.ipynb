```python
from bs4 import BeautifulSoup
from urllib.parse import urlparse
```

```python
import os
directory = 'urls'
dict1={}
for filename in os.listdir(directory):
    ##print(filename)
    fname='urls/'+filename
    f1=open(fname)
    j=filename.split('_')
    name=j[0]
    #print(name)
    soup1=BeautifulSoup(f1,'html.parser')
    result1=soup1.find_all('url')
    
    for k in range(0,len(result1)):
        try:
            unurl=result1[k].url.text
            #print(name,unurl)
            data = urlparse(unurl)
            murl=data.netloc
            dict1[name]=murl
            
        except:
            continue
for paper in dict1:
    print(paper,dict1[paper])
    
    
  
```
