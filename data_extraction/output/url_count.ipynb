```python
from bs4 import BeautifulSoup
```

```python
f=open("urls.txt",'r')
soup1=BeautifulSoup(f,'html.parser')
```

```python
dict_url={}
result1=soup1.find_all('url')
for k in range(0,len(result1)):
    try:
        furl=result1[k].url.text
        if len(furl) > 15:
            if furl in dict_url:
                dict_url[furl]+=1
            else:
                dict_url[furl]=1
        #print(furl)
    except:
        continue
```

```python
a1_sorted_keys = sorted(dict_url, key=dict_url.get, reverse=True)
dict1={}
for r in a1_sorted_keys:
    dict1[r]=dict_url[r]
```

```python
for k in dict1:
    print(k,dict1[k])
```
