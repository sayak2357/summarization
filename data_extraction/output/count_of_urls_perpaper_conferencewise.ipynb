```python
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import operator
import os
```

```python

directory = 'url_analysis/aclurl'
i=0
x=[]
y1=[]
y2=[]


for filename in os.listdir(directory):
    fname='url_analysis/aclurl/'+filename
    f1=open(fname)
    dict1={}
    c=0
    soup1=BeautifulSoup(f1,'html.parser')
    result1=soup1.find_all('url')
    #print(filename,len(result1))
    for k in range(0,len(result1)):
        unurl=result1[k].text
        data = urlparse(unurl)
        murl=data.netloc
        if len(murl)>6:
            c+=1
            if murl in dict1:
                dict1[murl]+=1
            else:
                dict1[murl]=1
    ct=0
    for j in dict1:
        ct+=dict1[j]
    x.append(i)
    y1.append(c)
    y2.append(ct)
    i+=1

pp=[]
for i in range(0,19):
    paper=0
    name='acl'+str(i)
    directory = 'pdfs/acl/'+name
    for filename in os.listdir(directory):
        paper+=1
    pp.append(paper)
for i in range(0,len(y1)):
    y1[i]=y1[i]/pp[i]
print(x)
print(y1)
print(y2)
print(pp)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects=x
y_pos = np.arange(len(objects))
performance = y1
plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('url count')
plt.xlabel('year (2000+) ')
plt.title('number of URLS per paper in ACL')
 
plt.show()
```
